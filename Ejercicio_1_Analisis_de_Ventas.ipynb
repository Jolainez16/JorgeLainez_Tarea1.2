```python
#Ejercicio 1 Analisis de Ventas
import pandas as pd
import numpy as np

#1.Cargar el archivo CSV en un DataFrame
df = pd.read_csv('./ventas.csv')
```

```python
#2.Calcular la cantidad total de productos vendidos por categoría
qty_categoria = df.groupby("Producto")["Cantidad"].sum()

print(qty_categoria)
```

```python
#3.Determine cuál es el producto con el mayor total de ventas

df['Total_Ventas'] = df['Cantidad'] * df['Precio_Unitario']

producto_venta = df.groupby('Producto')['Total_Ventas'].sum()

producto_mayor_venta = producto_venta.idxmax()

print("El producto con mayor total de ventas es:", producto_mayor_venta)
```

```python
#4.Encuentre el precio promedio de los productos vendidos

preciototal_por_producto = df.groupby('Producto')['Precio_Unitario'].sum()

cantidadtotal_por_producto = df.groupby('Producto')['Cantidad'].sum()

preciopromedio_por_producto = preciototal_por_producto / cantidadtotal_por_producto

promedio_produc = preciopromedio_por_producto.mean()

print("El precio promedio de los productos vendidos es:", promedio_produc)
print("Precio promedio por", preciopromedio_por_producto)

```
